```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your housing data (assuming you have a DataFrame)
# Replace 'your_data.csv' with the actual data source.
housing_data = pd.read_csv('your_data.csv')

# Set the style sheet
plt.style.use('your_preferred_stylesheet')

# Create a figure with the specified size
plt.figure(figsize=(12, 8))

# Create a line plot for ZN with a solid green line
plt.plot(housing_data['ZN'], label='ZN', color='green', linestyle='-')

# Create a line plot for INDUS with a blue dashed line
plt.plot(housing_data['INDUS'], label='INDUS', color='blue', linestyle='--')

# Add labels and a legend
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot of ZN and INDUS')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create dummy data
data = {'col1': [3, 5, 7, 9, 2],
        'col2': [4, 6, 8, 7, 1]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.bar(df.index, df['col1'], label='col1', color='blue')
plt.bar(df.index, df['col2'], label='col2', color='red', bottom=df['col1'])
plt.title('Bar Chart with Large Title')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend(loc='lower left')
plt.xticks(df.index, [f'Item {i}' for i in range(1, len(df) + 1)])
plt.show()

plt.figure(figsize=(8, 6))
plt.barh(df.index, df['col1'], label='col1', color='blue')
plt.barh(df.index, df['col2'], label='col2', color='red', left=df['col1'])
plt.title('Horizontal Bar Chart with Large Title')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.legend(loc='upper right')
plt.yticks(df.index, [f'Item {i}' for i in range(1, len(df) + 1)])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your data into a DataFrame (replace 'your_data.csv' with your data source)
data = pd.read_csv('your_data.csv')

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Find two variables with a strong positive correlation
# Replace 'col1' and 'col2' with the names of your correlated columns.
col1 = 'feature1'
col2 = 'feature2'

# Create a scatter plot for the identified variables
plt.figure(figsize=(8, 6))
plt.scatter(data[col1], data[col2, alpha=0.6, c='b', label=f'Correlation: {correlation_matrix.loc[col1, col2]:.2f}')

plt.title(f'Scatter Plot of {col1} vs. {col2}')
plt.xlabel(col1)
plt.ylabel(col2)
plt.legend(loc='best')
plt.grid(True)

plt.show()
```
